# HELLS WEEK DATASET MODEL USING TENSORFLOW

## IMPORTING LIBRARIES

```python
import numpy as np 
import pandas as pd                              # Numpy and Pandas for data handeling
import tensorflow as tf
from tensorflow import keras                      # Tensorflow and kereas for model building
from sklearn.preprocessing import MinMaxScaler    # For normalising dataset
```

## READING TRAINING AND TESTING DATA SETS

```python
df_train=pd.read_csv('train.csv')
df_test=pd.read_csv('test.csv')
```

## CREATING FUNCITION WHICH CAN EXTRACT NUMERIC VALUE IN THE STRING OF LAST 100 COLUMN OF TEST AND TRAIN DATA SET.

### EX:- 
#### INPUT:- '🍣🏿🟠U👨@-40.77392373961654@️bY‍t'
#### OUTPUT:- -40.77392373961654

```python
def value_extract(df,start_index,end_index):
    df.fillna(value='',inplace=True)              # Replacing all null values with ''
    for i in list(df)[start_index:end_index]:     # Finds the special character which is in strting and ending of number
        if df.loc[0,i]!='':                       # If string is '🍣🏿🟠U👨@-40.77392373961654@️bY‍t' the special char is @
            special_char=str(df.loc[0,i])[5]      # Every column has a special char and it is always at 5th index 
        else:                                     # If else conditional statement to check whether first string in column i is null or not
            x=0                                   # If it is not null then assign the character on 5th index as special char
            while(df.loc[0+x,i]==''):             # Otherwise run a while loop until which increases index untill string at that index is not null 
                x+=1
            special_char=str(df.loc[0+x,i])[5]
        df[i]=df[i].apply(lambda x: str(x)[str(x).find(special_char)+1:str(x).rfind(special_char)])    # Slices every entry and returns the number in string
    df=df.replace('','0.00')                                                                           # Replacing '' with '0'
    for i in list(df)[start_index:end_index]:    
        df[i]=df[i].astype(float)                                                                     # Changing the data type of every modified column to float
    return df
```

```python
df_train=value_extract(df_train,11,111)
df_test=value_extract(df_test,10,111)
```

```python
df_train.head()
```

## REORDERING THE COLUMN IN df_train WITH df_test

```python
df_train=pd.concat([df_train[list(df_test)],df_train['state']],axis=1)
```

## CREATING X_train,X_test BY DROPPING THE COLUMNS THAT ARE NOT NECESSARY LIKE:-UID,NAME ETC. AND CREATING Y_train BY ONE HOT ENCODING df_train['state']

```python
X_train=df_train.drop(list(df_train)[:10],axis=1)    # Droping UID,ph_no,cvv,credit_card_number,job,email,url,country,name because output is not dependent on them
X_train.drop(['state'],axis=1,inplace=True)          # Droping state because it is used for output
X_test=df_test.drop(list(df_test)[:10],axis=1)       # Droping UID,ph_no,cvv,credit_card_number,job,email,url,country,name because output is not dependent on them
Y_train=pd.get_dummies(df_train['state'])            # One hot encoding state to Y_train
```

## NORMALISING THE VALUES IN X_train AND X_test

```python
scaler=MinMaxScaler()                  
X_train=scaler.fit_transform(X_train)    # Normalising X_train value
X_test=scaler.fit_transform(X_test)      # Normalising Y_train value
```

## BUILDING AND TRAINING THE MODEL

```python
model=tf.keras.Sequential([
    keras.layers.Dense(100,input_shape=(100,),activation='elu'), # Input layer
    keras.layers.Dense(100,activation='elu'),                    # Hidden Layer #1
    keras.layers.Dense(100,activation='elu'),                    # Hidden Layer #2
    keras.layers.Dense(50,activation='elu'),                     # Hidden Layer #3
    keras.layers.Dense(50,activation='elu'),                     # Hidden Layer #4
    keras.layers.Dense(25,activation='elu'),                     # Hidden Layer #5
    keras.layers.Dense(25,activation='elu'),                     # Hidden Layer #6
    keras.layers.Dense(12,activation='elu'),                     # Hidden Layer #7
    keras.layers.Dense(12,activation='elu'),                     # Hidden Layer #8
    keras.layers.Dense(10,activation='sigmoid')                  # Output Layer
])                                                                                 # Creating model
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])    # Compiling Model
model.fit(X_train,Y_train,epochs=100)                                              # Training Model
```

## PREDICTING VALUE OF X TRAIN AND STORING IT ON TO prediction.csv

```python
Y_pred_raw=model.predict(X_test)                                  # Creating a 2D numpy array of ouput of all neuron
Y_pred=[]
for i in range(Y_pred_raw.shape[0]):
        Y_pred.append(list(Y_train)[np.argmax(Y_pred_raw[i])])    # Using outputs to genrating the prediction on basis of nueron which gave maximum value
Y_pred=pd.DataFrame({'state':Y_pred})                             # Converting list of prediction to dataframe
```

```python
prediction=pd.concat([df_test['UID'],Y_pred],axis=1)    # Concatinanting the prediction with UID 
prediction.to_csv('prediction.csv',index=False)         # Converting dataframe to csv
```
